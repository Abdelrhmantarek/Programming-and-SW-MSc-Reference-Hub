# Reading Files with Python (Text Files)
    open() function (read mode 'r') 
    read(), readline(), readlines() methods 

```python
#read text file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#write text file
with open("output.txt", "w") as file: 
		file.write("Hello, Python!")
```

```python
#read text file
with open("output.txt", "r") as file:
    content = file.read()
    print(content)
```

# Handling CSV Files
    Introduction to CSV files 
    csv.reader for reading 
    csv.writer for writing 
# csv -> comma-separated value
    column are separated by comma


```python
#read csv file
import csv
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
#write csv file
import csv
data = [
    ["Name", "Age", "City"],
    ["Alice", "30", "New York"],
    ["Bob", "25", "London"],
    ["Charlie", "35", "Paris"]
]
with open("output.csv", 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)  # Create a CSV writer object
    csvwriter.writerows(data)  
```

```python
#read csv file
import csv
with open("output.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

# Handling JSON Files
    
    json.load() for reading
    json.dump() for writing
# json -> JavaScript Object Notation
    data is set by key and value


```python
#write json file
import json
data = {"name": "Alice", "age": 30}

with open("data.json", "w") as f:
    json.dump(data, f)
```

```python
#read json file
import json
with open("data.json", 'r') as f:
           loaded_data = json.load(f)

print(loaded_data)
#read by key
print(loaded_data["name"])
```

# Handling Excel Files
    pandas.read_excel() for reading
    pandas.to_excel() for writing



```python
#write to excel
import pandas as pd
# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 28],
        'City': ['New York', 'London', 'Tokyo']}
df = pd.DataFrame(data)
df.to_excel('data.xlsx', index=False) #index=False prevents writing the index column
print("Excel file created successfully!")
```

```python
#read from excel
import pandas as pd
df = pd.read_excel("data.xlsx")
print(df.head())
```
