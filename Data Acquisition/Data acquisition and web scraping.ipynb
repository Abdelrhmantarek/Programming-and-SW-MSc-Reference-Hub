# Handling CSV Files
    Introduction to CSV files 
    csv.reader for reading 
    csv.writer for writing 
# csv -> comma-separated value
    column are separated by comma


```python
#read csv file
import csv
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
#write csv file
import csv
data = [
    ["Name", "Age", "City"],
    ["Alice", "30", "New York"],
    ["Bob", "25", "London"],
    ["Charlie", "35", "Paris"]
]
with open("output.csv", 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)  # Create a CSV writer object
    csvwriter.writerows(data)  
```

```python
#read csv file
import csv
with open("output.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

# Fetching Web Pages with Requests

```python
import requests
url = "https://www.airbnb.com/"
response = requests.get(url)
if response.status_code == 200:
    print("featched")
else:
    print(f"error{response.status_code}")
```

# Parsing HTML with Beautiful Soup

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content,"html.parser")
title = soup.title.text
print(f"Title: {title}")
```

# Finding Elements in HTML

```python

title_elements = soup.find_all(lambda tag: tag.has_attr('id') and tag['id'].startswith('title_'))

for element in title_elements:
    print(element.get_text())
```
